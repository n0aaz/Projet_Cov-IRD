```python
import plotly.graph_objects as go
import pandas as pd
from donnees import *

recuperer_stats_covid()

TableFrance= recup_pays('France')
TableDeces,TableInfection,TableDate= recup_champ(TableFrance,'Deces'),recup_champ(TableFrance,'Infection'),recup_champ(TableFrance,'Date')

coeff_opti=calcul_coeff(TableDeces,TableInfection)


fig = go.Figure()


fig.add_trace(
    go.Scatter(x=TableDate,
               y=TableInfection,
               name="Infectés cumulés",
               line=dict(color="#33CFA5")))

fig.add_trace(
    go.Scatter(x=TableDate,
               y=TableDeces,
               name="Deces",
               visible=False,
               line=dict(color="#33CFA5", dash="dash")))



fig.update_layout(
    updatemenus=[
        dict(
            active=0,
            buttons=list([
                dict(label="Infectés cumulés",
                     method="update",
                     args=[{"visible": [True, False, True, False]},
                           {"title": "Graphique",
                            "annotations": []}]),
                 dict(label="Deces",
                     method="update",
                     args=[{"visible": [True, True, False, False]},
                           {"title": "COVID-19"}]),
            ]),
        )
    ])



# Set title
fig.update_layout(title_text="COVID-19")


fig.show()

```
